```python
import pandas as pd
import matplotlib.pyplot as plt
filenames = ['', '', '', '', '']
def plot_csv_comparison(filenames):
    colors = ["#9a62c5", "#63aa5f", "#b49541", "#c55d8b", "#6596cd", "#c95b42",]
    max_values = []
    for filename, color in zip(filenames, colors):
        if filename:
            df = pd.read_csv(f'runs/train/VisDrone/{filename}/results.csv')
            x, y = df.iloc[:, 0], df.iloc[:, 7] * 100
            plt.plot(x, y, label=filename, color=color)
            max_y = y.max()
            max_x = x[y.idxmax()]
            #plt.scatter(max_x, max_y, color=color)  
            max_values.append((filename, max_x, max_y))
    for i, (filename, max_x, max_y) in enumerate(max_values):
        plt.text(1.02, 0.9 - i * 0.05, f'{filename}: Max mAP50 = {max_y:.2f}% at Epoch {max_x}', transform=plt.gca().transAxes, fontsize=9, color=colors[i])
    plt.xlabel('Epochs')
    plt.ylabel('mAP50')
    plt.title('Comparison of Model Training')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=9, frameon=False)
    plt.grid(True)
    plt.show()

filenames = [
    #'YOLO11n',
    #'YOLO11s',
    #'YOLO11m',
    #'YOLO11l',
    #'YOLO11x',
    
    'GOOD-Net-n',
    'GOOD-Net-s',
    'GOOD-Net-m',
    'GOOD-Net-l',

]
plot_csv_comparison(filenames)
```
